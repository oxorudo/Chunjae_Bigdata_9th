```python
import seaborn as sns
import matplotlib.pyplot as plt

# tips 데이터 로드
tips = sns.load_dataset("tips")

# pyplot을 활용한 line plot
# plt.plot()

# seaborn을 활용한 scatterplot
# scatterplot(data = dataframe, x = "열1", y = "열2")
sns.scatterplot(
    data=tips,
    x="total_bill",  # 청구 금액
    y="tip",  # tip 금액
)

plt.show()
```

```python
# 식사 인원수와 청구 금액의 관계를 산점도(scatter)로 표현하세요.

sns.scatterplot(
    data=tips,
    x="size",
    y="total_bill",
)
plt.show()
```

```python
# hue 인자 : 데이터(점)을 구별할 수 있음

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")  # 구분 기준 데이터
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 여러개의 axes에 seaborn 그래프 그리기
tips = sns.load_dataset("tips")
figs, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
    nrows=2,
    ncols=2,
)

# ax 인자 : 어떤 axes에 seaborn 그래프를 그릴지 결정
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    ax=ax1,
)

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    ax=ax2,
)

sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    ax=ax3,
)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# tips 데이터 로드
tips = sns.load_dataset("tips")

g = sns.FacetGrid(  # FaceGrid는 저장해줄 변수 필요
    data=tips,
    col="day",  # day 열을 기준으로 그래프를 시각화
    row="time",  # 식사 시간대 기준
)  # 날짜와 식사 시간대를 기준으로 그래프를 나누어 작성

g.map(
    sns.histplot,  # 히스토그램 sns.histplot
    "total_bill",  # total_bill에 대해 그래프를 작성
)

plt.show()
```

```python
# 행 기준 : time
# 열 기준 : smoker
# 출력 값 : tip

g = sns.FacetGrid(
    data=tips,
    row="time",
    col="smoker",
)  # FaceGrid는 저장해줄 변수 필요

g.map(
    sns.histplot,
    "tip",
)

plt.show()
```

```python
g = sns.FacetGrid(
    data=tips, row="time", col="smoker", hue="sex"  # hue는 여기에 추가 가능
)  # FaceGrid는 저장해줄 변수 필요

g.map(
    sns.scatterplot,  # scatterplot도 가능, x y 둘 다 존재해야 함
    "total_bill",
    "tip",
)

plt.show()
```
