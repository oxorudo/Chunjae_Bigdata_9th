```python
# 요소 탐색

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# 조건 설정을 위한 객체 불러오기
from selenium.webdriver.common.by import By

service = Service(
    ChromeDriverManager().install()
)  # 셀레니움이 크롬을 쓰기 위한 설치 과정

driver = webdriver.Chrome(service=service)  # 크롬 페이지를 여는 역할

url = "https://example.com"

# 웹 페이지 열기
driver.get(url)

# 단일 탐색, find_element(조건, 값)
p_tag = driver.find_element(By.TAG_NAME, "p")

# 복수 탐색, find_elements(조건, 값) : 똑같이 리스트에 저장됨
p_tags = driver.find_elements(By.TAG_NAME, "p")

for tag in p_tags:
    print(tag.text)

# print(p_tag.text)
# print(type(p_tag))

# driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

service = Service(
    ChromeDriverManager().install()
)  # 셀레니움이 크롬을 쓰기 위한 설치 과정

driver = webdriver.Chrome(service=service)  # 크롬 페이지를 여는 역할

url = "https://naver.com"
# 웹 페이지 열기
driver.get(url)

# id가 query인 요소를 탐색
# id의 By 조건 : By.ID
# find_element(조건,값)

# 입력 박스 탐색(접근)
search_box = driver.find_element(By.ID, "query")

# 입력 박스에 텍스트 입력
search_box.send_keys("python")

# 검색 버튼 탐색
button = driver.find_element(By.CLASS_NAME, "btn_search")

# 버튼 클릭 : click() 메서드 사용
button.click()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

service = Service(
    ChromeDriverManager().install()
)  # 셀레니움이 크롬을 쓰기 위한 설치 과정

driver = webdriver.Chrome(service=service)  # 크롬 페이지를 여는 역할

driver.implicitly_wait(10)  # 암시적 대기 : 모든 find_element에 적용됨
# 암시적 10초 대기 : 요소를 탐색할 때 10초까지 대기

url = "https://quotes.toscrape.com/"

# 웹 페이지 열기
driver.get(url)

quote_text_list = []

# 인용문과 인물을 감싼 div 복수 탐색
# -> class가 quote
quote_list = driver.find_elements(By.CLASS_NAME, "quote")


for quote in quote_list:
    quote_text = quote.find_element(By.CLASS_NAME, "text")
    quote_author = quote.find_element(By.CLASS_NAME, "author")

    quote_dic = {
        "quote": quote_text.text,
        "author": quote_author.text,
    }

    quote_text_list.append(quote_dic)

from pprint import pprint

pprint(quote_text_list)

driver.quit()
```
