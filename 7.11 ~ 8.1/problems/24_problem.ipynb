### (2) 공중화장실 표시하기

    - column 정보
        - 개방시간
        - 위도
        - 경도
1. Map 객체를 아래 인자로 생성합니다.
    - `location=[37.47875057229362, 126.88665978428628]`
    - `zoom_start=15`
2. 데이터프레임을 순회해서 Marker 객체를 생성하고 마커를 지도에 추가합니다.
    - 위도 경도 데이터를 활용해서 마커 위치를 지정하세요
    - 개방시간 데이터를 popup에 표시하세요.

```python
import folium

m = folium.Map(
    location=[37.5665, 126.9780],  # [위도,경도] : 서울시청 위치
    zoom_start=13,  # 최대 : 20, 최소 : 0
)

name = "마리오 아울렛 2관"
html = f"<b>{name}</b>"

# Popup 객체 생성
popup = folium.Popup(
    max_width=300,  # 최대 너비
    min_width=100,  # 최소 너비
    html=html,  # HTML Markup 문법 작성 인자
)

# Icon 객체 생성
icon = folium.Icon(
    icon="home",  # 아이콘 모양
    color="red",  # 아이콘 색
)

folium.Marker(
    location=[37.478, 126.886],
    popup=popup,
    icon=icon,  # icon 객체 지정
    tooltip="클릭",  # tooltip 객체 지정
).add_to(m)

m
```
