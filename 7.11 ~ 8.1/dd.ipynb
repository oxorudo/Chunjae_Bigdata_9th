```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

t = sns.FacetGrid(
    data=titanic,
    row="survived",
    col="embark_town",
    
    
)

t.map(
    sns.barplot,
    "pclass",
    "fare",
    order=[3, 2, 1], # barplot은 범주형 자료형을 지원하므로 자료형의 우선순위를 정해줘야 함
    label = "fare",
)

plt.legend()

plt.show()
```
