```python
import folium

m = folium.Map(
    location=[37.5665, 126.9780],  # [위도,경도] : 서울시청 위치
    zoom_start=15,  # 최대 : 20, 최소 : 0
)

m.save("map.html")
m
```

```python
# 마커 Marker 찍어보기

"""마커 객체 생성,인자
    folium.Marker(
    location='',
    popup="",
    tooltip=,
    icon=,
)"""

name = "서울역"
html = f"<b>{name}</b>"

# Popup 객체 생성
popup = folium.Popup(
    max_width=300,  # 최대 너비
    min_width=100,  # 최소 너비
    html=html,  # HTML Markup 문법 작성 인자
)

# Icon 객체 생성
icon = folium.Icon(
    icon="home",  # 아이콘 모양
    color="red",  # 아이콘 색
)


# marker 객체 생성
folium.Marker(
    location=[37.5665, 126.9780],
    popup="서울 시청",
    icon=icon, # icon 객체 지정
    tooltip="클릭" # tooltip 객체 지정
).add_to(
    m
)  # map 객체에 마커 추가

# ----------------------------------- 이하 두번째 마커(서울역) 생성

# Icon 객체 생성
icon = folium.Icon(
    icon="home",  # 아이콘 모양
    color="red",  # 아이콘 색
)

# marker 객체 생성
folium.Marker(
    location=[37.554, 126.970],
    popup=popup,
    icon=icon, # icon 객체 지정
    tooltip="클릭" # tooltip 객체 지정
).add_to(
    m
)  # map 객체에 마커 추가

m
```
