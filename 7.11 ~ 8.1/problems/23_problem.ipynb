```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

selected_diamonds = diamonds[["carat", "depth", "price"]]

corr_selected_diamonds = selected_diamonds.corr()

sns.heatmap(corr_selected_diamonds, annot=True)
plt.show()
```

```python
car = sns.load_dataset("car_crashes")

selected_car = car[["total", "speeding", "alcohol", "not_distracted", "no_previous"]]

corr_selected_car = selected_car.corr()

sns.heatmap(corr_selected_car, annot=True)
plt.show()
```

```python
labels = diamonds["cut"].unique()

size = diamonds["cut"].value_counts()

plt.pie(x=size, labels=labels, autopct="%1.1f%%")

plt.show()
```

```python
penguins = sns.load_dataset("penguins")

labels = penguins["species"].unique()

size = penguins["species"].value_counts()

plt.pie(x=size, labels=labels, autopct="%1.1f%%")
plt.show()
```
