### (1) 입금과 출금 그리고 조회

- 클래스(class) : BankAccount
- 인스턴스 변수(속성)
    - account_holder : 계좌 소유자
    - balance : 잔액
- 생성자 메서드(`__init__`)
    - account_holder : 계좌 생성자
    - balance : 잔액, 기본값 = 0
- 인스턴스 메서드
    - deposit(amount) : amount 만큼 잔액이 증가
    - withdraw(amount) : amount 만큼 잔액이 감소
- 베이스 코드 및 출력 결과

```python
class BankAccount:
		pass
		# pass 코드를 지우고 클래스 코드 작성
		
		

# 객체 생성 및 메서드 호출
account = BankAccount("홍길동")

# 잔액 조회
print(f"초기 잔액: {account.balance}원")

# 입금
account.deposit(1500)

# 출금
account.withdraw(300)

# 잔액 조회
print(f"현재 잔액: {account.balance}원")

account.withdraw(1200)

print(f"현재 잔액: {account.balance}원")

# 계좌 소유자 조회
print(f"계좌 소유자: {account.account_holder}")
```

```bash
초기 잔액: 0원
현재 잔액: 1200원
현재 잔액: 0원
계좌 소유자: 홍길동
```

````python
class BankAccount:

    def __init__(self, account_holder):
        self.account_holder = account_holder
        self.balance = 0

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def check(self):
        print(f"현재 잔액은 {self.balance}원 입니다.")


# 객체 생성 및 메서드 호출
account = BankAccount("홍길동")

# 잔액 조회
print(f"초기 잔액: {account.balance}원")

# 입금
account.deposit(1500)

# 출금
account.withdraw(300)

# 잔액 조회
# print(f"현재 잔액: {account.balance}원")
account.check()

account.withdraw(1200)

# print(f"현재 잔액: {account.balance}원")
account.check()

# 계좌 소유자 조회
print(f"계좌 소유자: {account.account_holder}")
````

### (2) 메서드 유효성 검사

> 프로그램에 입력된 값이 형식, 조건, 논리적으로 맞는지 확인하는 검사입니다.
입금과 출금 메서드에 유효성 검사를 추가해봅시다.
> 
- deposit(amount)
    1. 양의 정수 검사
        - amount가 숫자형 값인지 검사합니다.
        - amount가 양의 정수인지 검사합니다.
        - 검사를 통과하지 못하면 `잘못된 값입니다` 를 출력합니다.
- withdraw(amount)
    1. 양의 정수 검사
        - amount가 숫자형 값인지 검사합니다.
        - amount가 양의 정수인지 검사합니다.
        - 검사를 통과하지 못하면 `잘못된 값입니다` 를 출력합니다.
    2. 잔액 검사
        - amount가 잔액 보다 작거나 같은지 검사합니다.
        - 검사를 통과하지 못하면 `잔액이 부족합니다` 를 출력합니다.

- 베이스 코드 및 출력 결과
class BankAccount:
		pass
		# pass 코드를 지우고 클래스 코드 작성
		
		
 
```python
account = BankAccount("홍길동")
 
print(f"초기 잔액: {account.balance}원")

account.deposit("천오백원")

account.deposit(1500)

account.withdraw(-300)

print(f"현재 잔액: {account.balance}원")

account.withdraw(1600)

print(f"현재 잔액: {account.balance}원")
```

```bash
초기 잔액: 0원
잘못된 값입니다.
잘못된 값입니다.
현재 잔액 : 1500원
잔액이 부족합니다.
현재 잔액 : 1500원
```

````python
class BankAccount:

    def __init__(self, account_holder):
        self.account_holder = account_holder
        self.balance = 0

    def deposit(self, amount):
        try:
            if amount < 0:
                print("잘못된 값입니다.")
            else:
                self.balance += amount
        except TypeError as e:
            print("잘못된 값입니다.")

    def withdraw(self, amount):
        try:
            if amount < 0:
                print("잘못된 값입니다.")
            elif amount > self.balance:
                print("잔액이 부족합니다.")
            else:
                self.balance -= amount
        except TypeError as e:
            print("잘못된 값입니다.")


account = BankAccount("홍길동")

print(f"초기 잔액: {account.balance}원")

account.deposit("천오백원")

account.deposit(1500)

account.withdraw(-300)

print(f"현재 잔액: {account.balance}원")

account.withdraw(1600)

print(f"현재 잔액: {account.balance}원")
````
