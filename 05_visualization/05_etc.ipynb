```python
# 타이타닉 데이터간 상관관계 Heatmap : 정비례, 반비례 표시
# Heatmap : 수치형 자료

import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# 수치형 자료 추출
# print(titanic.info())
numeric_titanic = titanic.select_dtypes(include=["float64", "int64"])
# print(numeric_titanic.info())

# 상관 행렬 계산
corr_matrix = numeric_titanic.corr()

sns.heatmap(
    corr_matrix,
    annot=True,  # 그래프에 값 표시
)
plt.show()
```

```python
# 펭귄 데이터에 대해 상관관계 Heatmap

penguins = sns.load_dataset("penguins")

numeric_penguins = penguins.select_dtypes(include=["float64", "int64"])

corr_penguins = numeric_penguins.corr()

sns.heatmap(corr_penguins, annot=True)

plt.show()
```

```python
# 붓꽃(iris) 데이터 사용해보기
iris = sns.load_dataset("iris")

print(iris.info())

sns.pairplot(  # 자동으로 수치형 자료만 뽑아서 출력
    data=iris,
    hue="species",
    # vars=['sepal_length', 'sepal_width'], # 내가 보고싶은 것만 사용해서 그래프 출력
)
plt.show()
```

```python
# matplotlib 활용 pie chart
# plt.pie()
# 타이타닉 데이터
# 승객 등급(pclass)의 비율

import matplotlib.pyplot as plt
import seaborn as sns # 데이터 불러오기용으로 사용

titanic = sns.load_dataset("titanic")

'''
plt.pie(
    labels : 내부 요소의 이름
    x(or size) : 요소의 크기
)
'''
# 고유값
labels = titanic['pclass'].unique()

# 각 고유값의 개수
size = titanic['pclass'].value_counts()

plt.pie(
    x=size, labels=labels, autopct="%1.2f%%",
)
plt.show()
```
