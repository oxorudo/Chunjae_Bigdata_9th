```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# plot() : line plot
plt.plot(x, y, label="x - y")  # 라벨이 있어야 범례 표시 가능

# 그래프 제목(한글 불가)
plt.title("first line plot")

# x축, y축 라벨 설정
plt.xlabel("X_Axis")
plt.ylabel("Y_Axis")

# 범례 설정
plt.legend()

# 그래프 출력
plt.show()
```

```python
# figure Axes 생성해서 그래프 시각화

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Figure와 Axes 생성하는 함수 (figure: 전체 그래프의 컨테이너, Axes : 실제로 데이터가 시각화 되는 공간)
# plt.subplots()

# figure, axes 생성
fig, ax = plt.subplots()

# 실제로 그래프가 그려지는 공간 : Axes
ax.bar(x, y, label="x-y", color="g")  # 막대 그래프

# axes에 대한 레이아웃 설정
ax.set_xlabel("X_Axis")
ax.set_ylabel("Y_Axis")
ax.legend()

ax.set_title("Bar Plot Axes")

# 그래프 출력
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# nrow, ncol 인자로 생성할 axes 개수를 지정
# 1개의 figure, 4개의 axes 생성 : 위아래로 2개씩 그려짐
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)

# 첫 번째 (0,0) axes, line plot
ax1.plot(x, y)

# 두 번째 (0,1) axes, bor plot 코드 작성해보기
# label : y
# axes title : bar plot
# legend 설정
ax2.bar(x, y, label="y")

ax2.set_title("bar plot")
ax2.legend()

# 세 번째(1,0) axes, scatter plot
ax3.scatter(x,y)

# 굳이 4개 다 그릴 필요는 없음

# figure title
fig.suptitle("1 figure, 4 axes")

# 그래프 출력
plt.show()
```
