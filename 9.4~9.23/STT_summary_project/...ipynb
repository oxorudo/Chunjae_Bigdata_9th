```python
import os
import pandas as pd

# 텍스트 파일들이 저장된 폴더 경로
folder_path = "./ebs_script/new/"

# 미리 정한 태그 (여기서는 예시로 태그들을 나열) - 문자열로 변환
tags = {
    "1": "도형",
    "2": "도형",
    "3": "도형",
    "4": "도형",
    "5": "규칙성",
    "6": "측정",
    "7": "규칙성",
    "8": "측정",
    "9": "도형",
    "10": "측정",
    "11": "측정",
    "12": "측정",
    "13": "측정",
    "14": "도형",
    "15": "측정",
    "16": "측정",
    "17": "측정",
    "18": "측정",
    "19": "도형",
    "20": "도형",
    "21": "측정",
    "22": "측정",
    "23": "측정",
    "24": "측정"
}

# 폴더 내의 텍스트 파일을 순회하며 데이터프레임에 저장
for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):  # 텍스트 파일만 처리
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "r", encoding="utf-8") as file:
            text_content = file.read().strip()  # 파일 내용을 읽어옵니다.
        
        # 태그 지정 (파일명에 따른 태그를 저장, 숫자가 문자열로 처리)
        tag = tags.get(filename.replace(".txt", ""), "unknown")  # 확장자 제거 후 문자열로 비교
        
        # 새로운 데이터를 딕셔너리로 생성
        new_row = pd.DataFrame({
            'text': [text_content],
            'category': [tag.strip()],
            'source': ['ebs']
        })
        
        # 데이터프레임에 새로운 행 추가
        df = pd.concat([df, new_row], ignore_index=True)

```

```python
df = df.drop(columns=["Unnamed: 0"])
```
