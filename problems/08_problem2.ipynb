````python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time
from pprint import pprint
import json

# 검색 키워드(자유롭게 작성)
search_keyword = "남자%20슬랙스"
url = f"https://search.shopping.naver.com/search/all?query={search_keyword}"

# 스크롤 다운 입력 사이 대기 시간
scroll_pause_time = 0.5
# 스크롤 시도 횟수
max_scroll_attempts = 10

# 페이지 이동 횟수
page = 3

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)
driver.implicitly_wait(10)
driver.get(url)

actions = ActionChains(driver)

# 상품 정보 딕셔너리 저장할 리스트
item_list = []

# 아래에 코드를 작성하세요.

scroll_count = 15

pause_time = 0.5

for _ in range(scroll_count):
    # PAGE DOWN 키 누름 상호작용
    actions.send_keys(Keys.PAGE_DOWN).perform()

    # 키 누름 사이의 대기 시간
    time.sleep(pause_time)

items = driver.find_elements(By.CSS_SELECTOR, "div.product_item__MDtDF")

for i in range(0, page + 1):

    for _ in range(scroll_count):
        # PAGE DOWN 키 누름 상호작용
        actions.send_keys(Keys.PAGE_DOWN).perform()
        # 키 누름 사이의 대기 시간
        time.sleep(pause_time)

    items = driver.find_elements(By.CSS_SELECTOR, "div.product_item__MDtDF")

    for item in items:
        # 태그는 a 이고, 클래스 명은 "product_link__TrAac linkAnchor"
        item_name = item.find_element(
            By.CSS_SELECTOR, "a.product_link__TrAac.linkAnchor"
        )
        # 태그는 span 이고, 클래스 명은 "price_num__S2p_v"
        item_price = item.find_element(By.CSS_SELECTOR, "span.price_num__S2p_v")
        # price_delivery__yw_We
        item_delivery = item.find_element(By.CSS_SELECTOR, "span.price_delivery__yw_We")

        item_link = item_name.get_attribute("href")

        data = {
            "상품명": item_name.text,
            "가격": item_price.text,
            "주소": item_link,
            "배송료": item_delivery.text.replace("배송비\n", ""),
        }

        item_list.append(data)
    if i == page:
        break
    next_button = driver.find_element(By.CSS_SELECTOR, "a.pagination_next__pZuC6")
    next_button.click()


pprint(item_list)

with open("shopping_list.json", "w") as file:
    json.dump(item_list, file, indent=4, ensure_ascii=False)
````
