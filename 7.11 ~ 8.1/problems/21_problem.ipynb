```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

sns.histplot(
    data=titanic,
    x="fare",
    bins=15,
)
plt.show()
```

```python
sns.kdeplot(data=titanic, x="age", hue="survived", fill=True)
plt.show()
```

```python
tips = sns.load_dataset("tips")

sns.histplot(data=tips, x="total_bill", hue="time", multiple="stack", bins=15)

plt.show()
```

```python
sns.kdeplot(
    data=tips,
    x="tip",
    hue="smoker",
    fill=True,
)
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", y="tip", hue="time", kind="kde", rug=True)
plt.show()
```
