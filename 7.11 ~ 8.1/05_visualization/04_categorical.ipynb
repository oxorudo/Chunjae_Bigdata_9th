```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# bar plot
# 검은 막대기 : 신뢰구간
# 팁(tips) 데이터의 bar plot
# x = 요일, y= 팁 금액 시각화

tips = sns.load_dataset("tips")

sns.barplot(
    data=tips,
    x="day",
    y="tip",
)
plt.show()
```

```python
# 다이아몬드 데이터
# x: cut 품질, y : price 가격

diamonds = sns.load_dataset("diamonds")

sns.barplot(
    data=diamonds,
    x="cut",
    y="price",
)

plt.show()
```

```python
# 다이아몬드 데이터
# 컬러 color
# s.s.counterplot(y=)

sns.countplot(data=diamonds, y="color")

plt.show()
```

```python
# Box Plot

# 타이타닉 데이터
# x는 승객 등급 pclass
# y는 표 가격 fare

titanic = sns.load_dataset("titanic")

sns.boxplot(data=titanic, x="pclass", y="fare", hue="embarked")

plt.show()
```

```python
# boxen plot
# 팁 데이터
# x = day
# y = total_bill
# hue = time

tips = sns.load_dataset("tips")

sns.boxenplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="time",
)

plt.show()
```

```python
# violin plot
# 팁 데이터
# x = day
# y = total_bill
# hue = sex

sns.violinplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    inner="stick",
    split=False # 구분된 그래프를 나눌지 말지 결정(기본값 : False)
)
"""
inner에 대한 설명
box : 중앙값, 사분위수
quartile : 사분위수(Q1,Q2,Q3)
point : 데이터의 각 점
stick : 데이터의 선, 밀도 중시
"""
plt.show()
```
