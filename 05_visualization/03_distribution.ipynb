```python
import seaborn as sns
import matplotlib.pyplot as plt

# 펭귄 데이터에 대한 히스토그램
penguins = sns.load_dataset("penguins")

sns.histplot(
    data=penguins,
    x = "flipper_length_mm", # 날개 길이
    hue = "island", # 구분 기준 데이터
    # multiple = "dodge" # dodge = 각 범주(기준) 안 겹치게
    multiple = "stack" ,# 각 범주(기준) 쌓아서 배치
    # multiple = "fill" # 각 범주를 비율로 배치(0 ~ 1.0)
    bins=15, # bin(막대) 의 수
)

plt.show()
```

```python
# KDE Plot : 분포도로 보여줌
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

sns.kdeplot( 
    data = penguins,
    x="flipper_length_mm",
    hue = "island",
    fill = True, # 곡선 내부에 보기 쉽게 색을 채워 줌

)

plt.show()
```

```python
# 펭귄 데이터에서
# 무게 "body_mass_g" 의 분포도를 KDE plot 시각화
# 데이터 구부능 ㄴ펭귄 종 species
# 곡선 내부에 색을 채워주세요.

sns.kdeplot(
    data=penguins,
    x="body_mass_g",
    hue="species",
    fill=True,
)

plt.show()
```

```python
# Rug plot : 선분이 두꺼워질수록 그 지점에 많이 모여 있는 것
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

sns.rugplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="island",
)

plt.show()
```

```python
# displot() 활용해서
# rugplot + plot(histogram, kde plot)

# displot()은 분포형 plot 함수들의 상위 함수
sns.displot(
    data=penguins,
    x='flipper_length_mm', # x축 컬럼
    y="body_mass_g", # y축 컬럼
    hue="island", # 구분 컬럼
    kind="kde", # 메인 plot
    rug=True, # Rug plot 설정
    # fill=True, # 그래프 내부를 색칠
)

plt.show()
```
