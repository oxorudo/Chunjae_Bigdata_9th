```python
# 네이버 상품 목록 페이지에서 상품명을 추출 후 출력
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from pprint import pprint

# 웹 페이지 상호작용 객체 ActionChains
from selenium.webdriver.common.action_chains import ActionChains

# 키보드 객체
from selenium.webdriver.common.keys import Keys

service = Service(ChromeDriverManager().install())

driver = webdriver.Chrome(service=service)

# 대기 매커니즘 코드
driver.implicitly_wait(3)

keyword = "선풍기"

url = f"https://search.shopping.naver.com/search/all?query={keyword}"

driver.get(url)

# ActionChains 객체 생성
# WebDriver 객체를 인자로 전달해야 함
actions = ActionChains(driver)

# 스크롤 다운 횟수
scroll_count = 15

# 프로그램 대기를 위한 time 객체
import time

# 스크롤 다운 대기시간
pause_time = 0.5

# 스크롤 다운 횟수만큼 다운
for _ in range(scroll_count):
    # PAGE DOWN 키 누름 상호작용
    actions.send_keys(Keys.PAGE_DOWN).perform()

    # 키 누름 사이의 대기 시간
    time.sleep(pause_time)

items = driver.find_elements(
    By.CSS_SELECTOR, "div.product_item__MDtDF"
)  # 태그는 div 이고, 클래스 명은 "product_item__MDtDF"

for item in items:
    # 태그는 a 이고, 클래스 명은 "product_link__TrAac linkAnchor"
    item_name = item.find_element(By.CSS_SELECTOR, "a.product_link__TrAac.linkAnchor")
    # 태그는 span 이고, 클래스 명은 "price_num__S2p_v"
    item_price = item.find_element(By.CSS_SELECTOR, "span.price_num__S2p_v")

    print(item_name.text, ":", item_price.text)
```

```python
# 네이버 쇼핑창 크롤링 결과를 json에 저장해보자
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from pprint import pprint

# 웹 페이지 상호작용 객체 ActionChains
from selenium.webdriver.common.action_chains import ActionChains

# 키보드 객체
from selenium.webdriver.common.keys import Keys

service = Service(ChromeDriverManager().install())

driver = webdriver.Chrome(service=service)

# 대기 매커니즘 코드
driver.implicitly_wait(3)

keyword = "선풍기"

url = f"https://search.shopping.naver.com/search/all?query={keyword}"

driver.get(url)

# ActionChains 객체 생성
# WebDriver 객체를 인자로 전달해야 함
actions = ActionChains(driver)

# 스크롤 다운 횟수
scroll_count = 15

# 프로그램 대기를 위한 time 객체
import time

# 스크롤 다운 대기시간
pause_time = 0.5

# 스크롤 다운 횟수만큼 다운
for _ in range(scroll_count):
    # PAGE DOWN 키 누름 상호작용
    actions.send_keys(Keys.PAGE_DOWN).perform()

    # 키 누름 사이의 대기 시간
    time.sleep(pause_time)

items = driver.find_elements(
    By.CSS_SELECTOR, "div.product_item__MDtDF"
)  # 태그는 div 이고, 클래스 명은 "product_item__MDtDF"

# 딕셔너리를 저장할 리스트
data_list = []

for item in items:
    # 태그는 a 이고, 클래스 명은 "product_link__TrAac linkAnchor"
    item_name = item.find_element(By.CSS_SELECTOR, "a.product_link__TrAac.linkAnchor")
    # 태그는 span 이고, 클래스 명은 "price_num__S2p_v"
    item_price = item.find_element(By.CSS_SELECTOR, "span.price_num__S2p_v")

    data = {"상품명": item_name.text, "가격": item_price.text}

    data_list.append(data)

pprint(data_list)

# json 파일에 저장
with open("naver_shopping.json", "w") as file:
    json.dump(data_list, file, indent=4)
```
