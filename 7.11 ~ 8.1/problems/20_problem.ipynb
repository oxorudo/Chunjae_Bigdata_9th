````python
import seaborn as sns
import matplotlib.pyplot as plt


titanic = sns.load_dataset("titanic")

print(titanic.head())
````

````python
sns.scatterplot(
    data=titanic,
    x="age",
    y="fare",
)

plt.show()
````

````python
sns.scatterplot(
    data=titanic,
    x="age",
    y="fare",
    hue="survived",
)

plt.show()
````

````python
import seaborn as sns


diamonds = sns.load_dataset("diamonds")

print(diamonds.head())
````

````python
sns.scatterplot(
    data=diamonds,
    x="carat",
    y="price",
)
plt.show()
````

````python
sns.scatterplot(
    data=diamonds,
    x="carat",
    y="price",
    hue="color",
)
plt.show()
````

````python
sns.scatterplot(
    data=diamonds,
    x="carat",
    y="price",
    hue="cut",
)
plt.show()
````

````python
g = sns.FacetGrid(
    data=diamonds,
    col="cut",
)  # FaceGridÎäî Ï†ÄÏû•Ìï¥Ï§Ñ Î≥ÄÏàò ÌïÑÏöî

g.map(
    sns.scatterplot,  # scatterplotÎèÑ Í∞ÄÎä•, x y Îëò Îã§ Ï°¥Ïû¨Ìï¥Ïïº Ìï®
    "carat",
    "price",
)

plt.show()
````
