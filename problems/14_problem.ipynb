````python
import pandas as pd

df2 = pd.read_csv("data/titanic_2.csv")
df3 = pd.read_csv("data/titanic_3.csv")

df = pd.concat([df2, df3], ignore_index=True)

print(df.head())
print(df.tail())
print(df.info())

unique_df = df.drop_duplicates().copy()
print(unique_df)

unique_df.drop(columns=["alive", "embarked", "class", "deck", "who"], inplace=True)
print(unique_df.head())

unique_df.rename(
    columns={
        "survived": "생존 여부",
        "pclass": "티켓 등급",
        "sibsp": "동반한 형제자매 또는 배우자의 수",
        "parch": "동반한 부모 또는 자녀의 수",
        "fare": "티켓 요금",
        "embark_town": "탑승 항구",
        "adult_male": "성인 남성 여부",
        "alone": "동반자 여부",
    },
    inplace=True,
)

print(unique_df.head())

unique_df.loc[unique_df["age"] < 1, "age"] = 1

print(unique_df.sort_values(by="age")["age"])

f1 = unique_df.loc[unique_df["티켓 등급"] == 1, "티켓 요금"].mean()
f2 = unique_df.loc[unique_df["티켓 등급"] == 2, "티켓 요금"].mean()
f3 = unique_df.loc[unique_df["티켓 등급"] == 3, "티켓 요금"].mean()

unique_df.loc[
    (unique_df["티켓 등급"] == 1) & (unique_df["티켓 요금"] == 0), "티켓 요금"
] = f1
unique_df.loc[
    (unique_df["티켓 등급"] == 2) & (unique_df["티켓 요금"] == 0), "티켓 요금"
] = f2
unique_df.loc[
    (unique_df["티켓 등급"] == 3) & (unique_df["티켓 요금"] == 0), "티켓 요금"
] = f3

print(unique_df.sort_values(by="티켓 요금")[['티켓 등급','티켓 요금',]])

print(unique_df["age"].mean())

print(unique_df["티켓 등급"].mode())

print(unique_df["탑승 항구"].unique())

print(unique_df["탑승 항구"].value_counts())

print(unique_df.loc[unique_df["age"] < 20, ["생존 여부", "age"]])

print(
    unique_df.loc[(unique_df["age"] < 20) & (unique_df["생존 여부"] == 1), "티켓 등급"]
)
````
