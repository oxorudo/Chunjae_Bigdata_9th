```python
from transformers import pipeline
from PIL import Image
# from googletrans import Translator
import matplotlib.pyplot as plt
import sentencepiece

# 오류가 난다면 파이썬 3.8버전으로 시도해보자.
```

```python
pipe = pipeline("document-question-answering", model="naver-clova-ix/donut-base-finetuned-docvqa")
# 영어만 지원됨.
```

```python
document = Image.open('./2.jpg')
```

```python
plt.imshow(document)
plt.axis(False)
plt.show()
```

```python
# 질문 입력 및 답변 도출
question = 'How much is Heineken?'
answer = pipe(image=document, question=question)
print(answer)
```
