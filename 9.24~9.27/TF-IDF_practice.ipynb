```python
# 조사 목록 정의
josa_list = [
    # 주격 조사
    '이', '가',

    # 목적격 조사
    '을', '를',

    # 관형격 조사
    '의',

    # 부사격 조사
    '에', '에서', '에게', '한테', '로', '으로', '까지', '부터', '와', '과', '보다',

    # 호격 조사
    '여', '이여',

    # 인용격 조사
    '라고', '고',

    # 접속 조사
    '와', '과', '하고', '이랑',

    # 보조사
    '은', '는', '도', '만', '밖에', '까지', '조차', '마저', '마는', '이나', '나', '든지', '든가', '나마', '마저', '라야', '뿐', '이라도',

    # 비교 조사
    '보다', '처럼', '같이', '만큼',

    # 기타 조사
    '께', '께서', '라도', '야말로', '으로서', '으로써', '이야말로', '이나마', '하고는', '마다','으로'
    '이', '가', '은', '는', '을', '를', '에', '와', '과', '로', '으로', 
    '의', '에서', '까지', '도', '에게', '한테', '부터', '만큼', '보다',
    '으로부터', '하고', '밖에', '마저', '조차', '라도', '께', '께서'
]
```

```python
def remove_josa(text): # 조사 제거 함수
    
    # 문자열을 공백 기준으로 word별로 나눕니다.
    word_list = text.split()
    
    # word에서 조사 제거
    result = []
    for word in word_list:
        for josa in josa_list:
            # 단어가 josa로 끝나면 해당 josa를 제거
            if word.endswith(josa):
                word = word[:-len(josa)]
                break  # 하나의 조사만 제거하고 멈춤
        result.append(word)
    
    # 다시 공백으로 연결
    return ' '.join(result)

# 테스트
문장 = "사람이 책을 읽는다"
print(remove_josa(문장))
```

```python
text_removed_josa = []
# 문단마다 조사 제거하고 다시 리스트에 저장
for sentence in text:
    text_removed_josa.append(remove_josa(sentence))
```

```python
# 조사 제거된 텍스트
text_removed_josa
```

```python
arr = vector.fit_transform(text_removed_josa).toarray()
```

```python
col = vector.get_feature_names_out()
```

```python
df = pd.DataFrame(arr, columns=col)
```

```python
# 문단별 가장 중요한 키워드
df.idxmax(axis=1)
```
